```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sale = [22,23,44,56,78,45]
months = ['jan','feb','mar', 'apr', 'may', 'jun']
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
plt.plot(months, sale, color='green', linestyle='--', marker ='o', label='2025 sale data ')
plt.title('Sales Report')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
plt.grid(color='gray', linestyle=':')
plt.xlim(0,6)
plt.ylim(20,80)
plt.show()
```

```python
product=['A','B', 'C', 'D', 'E' , 'F']
sales = [100, 1200,100,500,900,600]

plt.bar(product, sales ,color = 'red', label='2025 frist monts sale')

plt.title('Sales Report')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.legend()
plt.grid(color='gray', linestyle=':')
plt.show()
```

```python
direction = ['north', 'east','west','south']
production = [1000, 100,1500,650]

plt.pie(production, labels=direction, autopct='%1.1f%%',colors=['red', 'green', 'yellow' , 'pink'])
plt.title('sale Report')
plt.show()
```

```python
x = [1,4,6,8]

y = [100,50,80,60]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('line chart')
plt.grid(True)

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('bar chart')
plt.grid(color='gray', linestyle=':')

plt.show()
```

```python
product = ['a', 'b', 'c', 'd', 'e']
price = [100,300,50,150,250]

fig, ax = plt.subplots(1,2,figsize=(10,5))
ax[0].plot(product, price)
ax[0].set_title('line chart')
ax[0].grid()

ax[1].bar(product, price)
ax[1].set_title('bar chart')
ax[1].grid(color='gray', linestyle=':')

plt.tight_layout()
plt.show()
plt.savefig('product_sale.png', dpi = 1200 , bbox_inches='tight')
```
