```python
def extract_episode(text):
    check = False
    data = ""
    for i in text:
        if i == '(':
            check = True
            continue
        if i == ')':
            check = False
            break
        if check:
            data += i
    return data
```

```python
df["Episode"] = df["Title"].apply(extract_episode)
```

```python
import re

def extract_time_period(text):
    # Search for pattern: (Month Year - Month Year)
    match = re.search(r'([A-Za-z]{3} \d{4} - [A-Za-z]{3} \d{4})', text)
    if match:
        return match.group(1)
    else:
        return None

```

```python
df["time_priod"] = df["Title"].apply(extract_time_period)
```

```python
df
```

```python
from datetime import datetime

def calculate_total_months(time_period):
    try:
        # Split start and end
        start_str, end_str = time_period.split(' - ')

        # Convert to datetime
        start_date = datetime.strptime(start_str.strip(), '%b %Y')
        end_date = datetime.strptime(end_str.strip(), '%b %Y')

        # Calculate difference in months
        total_months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month) + 1

        return total_months
    except:
        return None  # In case of missing or bad data

```

```python
 df['time_priod'].apply(calculate_total_months)
```

```python
df['Total Months'] = df['time_priod'].apply(calculate_total_months)
```
